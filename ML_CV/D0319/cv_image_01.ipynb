```python
## 모듈 로딩 및 데이터 진부
import cv2          # open cv
import numpy as np
import matplotlib.pyplot as plt
```

```python
img_path = '../_data/_image/'
```

```python
##[2] 이미지 로딩
file = img_path+'20250112144858_1.jpg'

myImg = cv2.imread(file)
```

```python
print(f"형태: {myImg.shape}")
print(f"차원: {myImg.ndim}D")
print(f"타입: {myImg.dtype}")
```

```python
## 이미지출력
plt.imshow(myImg)
plt.title('myImg')
plt.show()
```

```python
## 이미지 팝업 출력창
# 이미지 창 활성화 후 아무 키나 입력하면 종료됨.
cv2.imshow("[myImg]", myImg)
#이미지 쇼
cv2.waitKey(10000)
# 키보드 입력대기 (시간)
cv2.destroyAllWindows()
# 입력시 발생 이벤트
```

```python
import seaborn as sns
import pandas as pd # 데이터 분석 및 전처리
import numpy as np # 숫자처리
import matplotlib.pyplot as plt # 데이터 시각화
import cv2
import os
```

```python
imgDir = '../_data/_image/'
videoDir = '../_data/_video/'
fileName = 'Dog_ (57).jpg'
```

```python
colorImg = cv2.imread(imgDir+fileName, cv2.IMREAD_COLOR)
if colorImg is not False:
    print(fileName)
    print(f"shape: {colorImg.shape}     ndim: {colorImg.ndim}")
else:
    print(f'{colorImg} check')
```

3-1 이미지 축소
     이미지데이터 , 축소w, h


```python
## 이미지 확인
rgbIMG = cv2.cvtColor(colorImg, cv2.COLOR_BGR2RGB)

plt.imshow(rgbIMG)
plt.grid()
plt.show()
```

[3] 이미지자르기

```python
## 현재 이미지 크기 
H, W, _ = rgbIMG.shape

## - 자르고 싶은 크기
X, Y, nh, nw =  500,200, 250, 250
## 가능한 범위 여부 체크.
isValid =  (X+nw <= W) and (Y+nh <= H)
print(f"가능한 범위 여부{isValid}")
cropIMG = rgbIMG[Y:Y+nh,X:X+nw]
```

```python
plt.imshow(rgbIMG)
plt.imshow(cropIMG)
plt.show()
```

```python
cropIMG = cv2.cvtColor(cropIMG, cv2.COLOR_BGR2RGB)
```

[3-1] 이미지 대칭

```python
## 상하대칭
top_dwonImg = cv2.flip(colorImg, 0)
plt.imshow(cv2.cvtColor(top_dwonImg, cv2.COLOR_BGR2RGB))
plt.show()
```

```python
## 좌우대칭
hflipImg = cv2.flip(colorImg, 1)
plt.imshow(cv2.cvtColor(hflipImg, cv2.COLOR_BGR2RGB))
plt.show()
```

[4] 이미지 회전
- 회전행렬 추출 단계 : getRotationMatrix2D(중심좌표, 회전 각도(60), 확대/축소 배율)
- 변화단계 : warpAffine(원본이미지, 회전행렬, (행,열))


```python
rgbIMG.shape
```

```python
## [4-1] 회전행렬 추출
H, W, _ = rgbIMG.shape
# rotate45 = cv2.getRotationMatrix2D((H/2, W/2), 45, 0.5)
# rotate45 = cv2.getRotationMatrix2D((W/2,H/2), 45, 1)
rotate45 = cv2.getRotationMatrix2D((W/2,H/2), 45, 1)

img45 = cv2.warpAffine(rgbIMG, rotate45, (W, H))

plt.imshow(img45)
```

#### [ 이미지프로세싱 기초 - 스레시홀딩 ]

```python
## 1. 모듈 로딩및 데이터 준비
import cv2
        # 일반적인 이미지 파일들에서 
        # 순수 이미지 데이터 다루는 모듈
import matplotlib.pyplot as plt
            # 이미지 시각화
import numpy as np 
            ## 순수 이미지 데이터 저장 형식이 nadrray
import os   ## 폴더 파일 경로 관련
```

```python
img = cv2.imread('../_data/_image/cat_(61).jpeg', cv2.IMREAD_GRAYSCALE)

thresh_np  = np.zeros_like(img)
thresh_np[ img > 127] = 255
# print(img)
ret, thresh_cv =  cv2.threshold(img, 127, 255, cv2.THRESH_BINARY)
print(ret)

imgs = {'Origin':img, 'Numpy Api': thresh_np, 'cv2.threshold': thresh_cv}
for i, (key, value) in enumerate(imgs.items()):
    plt.subplot(1, 3, i+1)
    plt.title(key)
    plt.imshow(value, cmap='gray')
    plt.xticks([]); plt.yticks([])
plt.show()
```

```python
_, t_bin = cv2.threshold(img,127,255, cv2.THRESH_BINARY)
_, t_bininv = cv2.threshold(img,127,255, cv2.THRESH_BINARY_INV)
_, t_trunc = cv2.threshold(img,127,255, cv2.THRESH_TRUNC)
_, t_2zr = cv2.threshold(img,127,255, cv2.THRESH_TOZERO)
_, t_2zrinv = cv2.threshold(img,127,255, cv2.THRESH_TOZERO_INV)

imgs = {'origin':img, 'BINARY':t_bin, 'Binary_INV':t_bininv,
        'TRUNC':t_trunc, 'TOZERO':t_2zr, 'TOZERO_INV':t_2zrinv}
for i, (key, value) in enumerate(imgs.items()):
    plt.subplot(2,3, i+1)
    plt.title(key)
    plt.imshow(value, cmap='gray')
    plt.xticks([]); plt.yticks([])
plt.show()
```

#### [ 이미지를 파일 데이터셋으로 생성 ]
- 이미지 파일의 raw data 사용
- 데이터셋 파일 : csv, json, ... 등등 개발자 선택
- 데이터셋 파일 구성 : 라벨 + 피쳐, 피쳐 +라벨.
- 1개의 행(row)에 한 개 이미지가 저장.


```python
## 1. 모듈 로딩및 데이터 준비
import cv2
        # 일반적인 이미지 파일들에서 
        # 순수 이미지 데이터 다루는 모듈
import matplotlib.pyplot as plt
            # 이미지 시각화
import numpy as np 
            ## 순수 이미지 데이터 저장 형식이 nadrray
import os   ## 폴더 파일 경로 관련
```

```python
## [1-2] 데이터준비
IMG_DIR = '../_data/_image/cat_dog_copy/'
DS_DIR = '../_data/_image/cat_dog_color/'
ROW, COL = 60, 60
IMG_CSV = '../_data/animal_color.csv' ## 이미지 라벨과 이미지 로우데이터 저장 파일
```

```python
## 학습용 이미지 저장 폴더 생성
if not os.path.exists(DS_DIR):
    os.mkdir(DS_DIR)
else:
    print(f"{DS_DIR} already existed")

```

```python
IMG_FILELIST = os.listdir(IMG_DIR)

IMG_FILELIST = list(filter(lambda filename: os.path.isfile(IMG_DIR + filename), IMG_FILELIST))

# if DEBUG_MODE: print(IMG_FILELIST)
   
```

[2] 이미지 데이터 로딩

```python
## rgbsclae  형식으로 이미지 데이터 로딩
with open(IMG_CSV, mode='a') as csvF:
    for filename in IMG_FILELIST[:5]:
        # - 3채널로 이미지 로딩
        rgbIMG = cv2.imread(IMG_DIR+filename, cv2.IMREAD_COLOR)
        print(f"rgbIMG { rgbIMG.shape}")
        
        # - 사이즈 조정
        rgbIMG = cv2.resize(rgbIMG, (ROW, COL))
        print(f"rgbIMG => {rgbIMG.shape}")
        
        #- 이미지 저장
        resizeIMG = f"{DS_DIR}rgb_{filename}"
        print(f"resizeIMG => {resizeIMG}")
        ret = cv2.imwrite(resizeIMG, cv2.cvtColor(rgbIMG, cv2.COLOR_BGR2RGB) )
        if not ret: print(f"Fail {rgbIMG}")
        
        ## - 채널 분리
        b,g,r, = cv2.split(rgbIMG)    
        b, g, r = b.reshape(-1), g.flatten(), r.reshape(-1)
        label = filename[:3]
        data = np.concat((r,g,b))
        print(f"label {label}, data => {data.shape}")
        
        # # - csv 파일 쓰기 -> 라벨, 픽셀1, 픽셀2, 픽셀3, ...
        # label = filename[:3]
        # data = rgbIMG.reshape(-1)
        # print(f"label {label}, data => {data.shape}")
        sdata = list(map(lambda n: str(n), data))
        print(f"sdata => {sdata[:30]}")
        
        csvF.write(str(label)+',')          #숫자 라벨 쓰기
        csvF.write(','.join(sdata) + '\n')  #리스트 이미지 데이터 => 문자열 변환
```

```python

# Chrome 웹드라이버 설정
driver = webdriver.Chrome(service=Service(ChromeDriverManager().install()))

# CarWale 사이트 열기
driver.get("https://www.carwale.com/")

# 크롤링할 차량 리스트 (예시)
car_keywords = keywordList

for idx, car in enumerate(car_keywords, start=1):
    try:
        # 검색창 찾기 및 입력
        search_box = WebDriverWait(driver, 10).until(
            EC.presence_of_element_located((By.XPATH, '//input[@placeholder="Search"]'))
        )
        search_box.clear()
        time.sleep(1)
        search_box.send_keys(car)
        time.sleep(1)
        search_box.send_keys(Keys.RETURN)
        time.sleep(3)

        # 검색 결과 확인
        search_results = driver.find_elements(By.XPATH, '//div[@class="results-container"]//a[@class="car-card__link"]')

        if not search_results:
            print(f"🚨 검색 결과 없음: {car}")
            car_prices[car] = "검색 결과 없음"
            continue

        # 첫 번째 차량 클릭
        search_results[0].click()
        time.sleep(3)

        # 가격 정보 추출
        try:
            price_element = WebDriverWait(driver, 10).until(
                EC.presence_of_element_located((By.XPATH, ' '))
            )
            car_prices[car] = price_element.text
            print(f"✅ {idx}. {car}: {car_prices[car]}")
        except Exception:
            print(f"⚠️ 가격 정보 없음: {car}")
            car_prices[car] = "가격 정보 없음"

        # 뒤로 가기
        driver.back()
        time.sleep(2)

    except Exception as e:
        print(f"❌ Error occurred for {car}: {e}")
        car_prices[car] = "크롤링 실패"
        driver.back()
        time.sleep(2)

# 📌 크롤링 결과 출력
print("\n📌 크롤링 결과:")
for car, price in car_prices.items():
    print(f"{car}: {price}")

# 브라우저 종료
driver.quit()
```

```python

# Chrome ÏõπÎìúÎùºÏù¥Î≤Ñ ÏÑ§Ï†ï
driver = webdriver.Chrome(service=Service(ChromeDriverManager().install()))

# CarWale ÏÇ¨Ïù¥Ìä∏ Ïó¥Í∏∞
driver.get("https://www.carwale.com/")

# ÌÅ¨Î°§ÎßÅÌï† Ï∞®Îüâ Î¶¨Ïä§Ìä∏ (ÏòàÏãú)
car_keywords = keywordList

for idx, car in enumerate(car_keywords, start=1):
    try:
        # Í≤ÄÏÉâÏ∞Ω Ï∞æÍ∏∞ Î∞è ÏûÖÎ†•
        search_box = WebDriverWait(driver, 10).until(
            EC.presence_of_element_located((By.XPATH, '//input[@placeholder="Search"]'))
        )
        search_box.clear()
        time.sleep(1)
        search_box.send_keys(car)
        time.sleep(1)
        search_box.send_keys(Keys.RETURN)
        time.sleep(3)

        # Í≤ÄÏÉâ Í≤∞Í≥º ÌôïÏù∏
        search_results = driver.find_elements(By.XPATH, '//div[@class="results-container"]//a[@class="car-card__link"]')

        if not search_results:
            print(f"üö® Í≤ÄÏÉâ Í≤∞Í≥º ÏóÜÏùå: {car}")
            car_prices[car] = "Í≤ÄÏÉâ Í≤∞Í≥º ÏóÜÏùå"
            continue

        # Ï≤´ Î≤àÏß∏ Ï∞®Îüâ ÌÅ¥Î¶≠
        search_results[0].click()
        time.sleep(3)

        # Í∞ÄÍ≤© Ï†ïÎ≥¥ Ï∂îÏ∂ú
        try:
            price_element = WebDriverWait(driver, 10).until(
                EC.presence_of_element_located((By.XPATH, ' '))
            )
            car_prices[car] = price_element.text
            print(f"‚úÖ {idx}. {car}: {car_prices[car]}")
        except Exception:
            print(f"‚ö†Ô∏è Í∞ÄÍ≤© Ï†ïÎ≥¥ ÏóÜÏùå: {car}")
            car_prices[car] = "Í∞ÄÍ≤© Ï†ïÎ≥¥ ÏóÜÏùå"

        # Îí§Î°ú Í∞ÄÍ∏∞
        driver.back()
        time.sleep(2)

    except Exception as e:
        print(f"‚ùå Error occurred for {car}: {e}")
        car_prices[car] = "ÌÅ¨Î°§ÎßÅ Ïã§Ìå®"
        driver.back()
        time.sleep(2)

# üìå ÌÅ¨Î°§ÎßÅ Í≤∞Í≥º Ï∂úÎ†•
print("\nüìå ÌÅ¨Î°§ÎßÅ Í≤∞Í≥º:")
for car, price in car_prices.items():
    print(f"{car}: {price}")

# Î∏åÎùºÏö∞Ï†Ä Ï¢ÖÎ£å
driver.quit()
```

```python
keywordList = []
with open('keyword', 'r') as f:
    text = f.readlines()
    for i in text:
        keywordList.append(i.strip())
# for i in range(len(keywordList)):
#     keywordList[i] = keywordList[i]
```

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.chrome.service import Service
from webdriver_manager.chrome import ChromeDriverManager
import time

# Chrome  
driver = webdriver.Chrome(service=Service(ChromeDriverManager().install()))

# CarWale  
driver.get("https://www.carwale.com/")

#      (  )
car_keywords = keywordList  # 

car_prices = {}
count=0
time.sleep(5)
for car in car_keywords:
    try:
        #   
        count+=1
        search_box = driver.find_element(By.XPATH, '//input[@placeholder="Search"]')
        search_box.clear()
        search_box.send_keys(car)
        time.sleep(1)
        search_box.send_keys(Keys.RETURN)
        time.sleep(5)

        # 3 XPath   
        x_paths = [
            '//span[@class="o-cJrNdO o-byFsZJ o-bkmzIL o-bVSleT"]',  #   
            '//span[contains(@class, "o-Hyyko") and contains(@class, "o-bPYcRG") and contains(@class, "o-eqqVmt")]',  #   
            '//*[@id="root"]/div[2]/div/div/section/div/ul[1]/li[1]/div/div/div/div/div[1]/span'  #   
        ]
        
        price = None
        for x_path in x_paths:
            try:
                
                price = driver.find_element(By.XPATH, x_path).text
                if price:  #    
                    break
            except Exception as e:
                
                continue  #     XPath 
        
        if price:
            print(count, car, price)
            car_prices[car] = price
        else:
            print(count, 'false')
            car_prices[car] = "Price not found"
        
    except Exception as e:
        print(f"Error occurred for {car}: {e}")
        car_prices[car] = "Error occurred"

#    
for car, price in car_prices.items():
    print(f"{car}: {price}")

#  
driver.quit()
```

```python
with open('car_p2.txt', 'w') as f:
    for key, value in car_prices.items():
        f.write(f"{key}: {value}\n")
```

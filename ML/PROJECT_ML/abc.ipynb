```python
keywordList = []
with open('keyword', 'r') as f:
    text = f.readlines()
    for i in text:
        keywordList.append(i.strip())
# for i in range(len(keywordList)):
#     keywordList[i] = keywordList[i]
```

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.chrome.service import Service
from webdriver_manager.chrome import ChromeDriverManager
import time

# Chrome 웹드라이버 설정
driver = webdriver.Chrome(service=Service(ChromeDriverManager().install()))

# CarWale 사이트 열기
driver.get("https://www.carwale.com/")

# 예시로 차량 키워드 리스트를 제공 (필요에 맞게 수정)
car_keywords = keywordList  # 예시

car_prices = {}
count=0
time.sleep(5)
for car in car_keywords:
    try:
        # 검색창에 키워드 입력
        count+=1
        search_box = driver.find_element(By.XPATH, '//input[@placeholder="Search"]')
        search_box.clear()
        search_box.send_keys(car)
        time.sleep(1)
        search_box.send_keys(Keys.RETURN)
        time.sleep(5)

        # 3가지 XPath로 요소 추출 시도
        x_paths = [
            '//span[@class="o-cJrNdO o-byFsZJ o-bkmzIL o-bVSleT"]',  # 첫 번째 경우
            '//span[contains(@class, "o-Hyyko") and contains(@class, "o-bPYcRG") and contains(@class, "o-eqqVmt")]',  # 두 번째 경우
            '//*[@id="root"]/div[2]/div/div/section/div/ul[1]/li[1]/div/div/div/div/div[1]/span'  # 세 번째 경우
        ]
        
        price = None
        for x_path in x_paths:
            try:
                
                price = driver.find_element(By.XPATH, x_path).text
                if price:  # 가격이 추출되면 멈추고 나감
                    break
            except Exception as e:
                
                continue  # 요소를 찾지 못하면 다음 XPath로 넘어감
        
        if price:
            print(count, car, price)
            car_prices[car] = price
        else:
            print(count, 'false')
            car_prices[car] = "Price not found"
        
    except Exception as e:
        print(f"Error occurred for {car}: {e}")
        car_prices[car] = "Error occurred"

# 크롤링한 차량 가격 출력
for car, price in car_prices.items():
    print(f"{car}: {price}")

# 브라우저 종료
driver.quit()
```

```python
with open('car_p2.txt', 'w') as f:
    for key, value in car_prices.items():
        f.write(f"{key}: {value}\n")
```
